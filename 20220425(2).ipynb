Auto correlation

(Meaning of auto correlation for sinal)

Definition: A correlation that exists between a series of observations that are sequential in time or space.

Intended use:

Autocorrelation analysis is to find repeating patterns such as the presence of periodic signals obscured by noise.
A mathematical tool for identifying the missing fundamental frequencies in signals implied by harmonic frequencies
In particular, signal processing provides spectrum information of signals by showing a series of values ​​such as domain signals.
* In time series data, the present state is often closely related to the past and future states. In this case, it is said that it is related to autocorrelation.

Root mean square (RMS) - Root mean square (RMS)

Definition: The positive square root (rooted) of the average of the squares of each variable, that is, the root mean square.

characteristic:

It is similar to the average value and serves as a representative value.
This is a useful method when you want to find out the degree of magnitude when the distribution of variables has positive and negative values ​​evenly.

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa, librosa.display
```

```python
# from google.colab import files
# fn = files.upload()
import os
url = "https://raw.githubusercontent.com/hsnam95/class2022Spring/main/aeiou.wav"
os.system("curl " + url + " > aeiou.wav")

s, sr = librosa.load('aeiou.wav')
```

```python
s = librosa.util.normalize(s)
librosa.display.waveplot(s, sr)
ipd.Audio(s[7000:12000], rate=sr)
```

```python
from scipy.signal import lfilter
sig = s[7000:12000]
sig = lfilter(np.array([1]), np.array([1]), sig, axis=0)
librosa.display.waveplot(sig, sr)
ipd.Audio(sig, rate=sr)
```

```python
from scipy.signal import lfilter
sig = s[7000:12000]
sig = lfilter(np.array([1/2, 1/2, 1/2]), np.array([1/2, -1/2]), sig, axis=0)
librosa.display.waveplot(sig, sr)
ipd.Audio(sig, rate=sr)
```
