```python
import math
y_actual = [1,2,3,4,5]
y_predicted = [1.6,2.5,2.9,3,4.1]
```

```python
MSE = np.square(np.subtract(y_actual,y_predicted)).mean()
```

```python
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
import numpy as np
```

```python
def window_rms(a, window_size):
  a2 = np.power(a,2)
  window = np.ones(window_size)/float(window_size)
  return np.sqrt(np.convolve(a2, window, 'valid'))
```

Y(t)=(a*SIN(2*PAI*f*t))

```python
# amplitude: 325.27 volt is the peak voltage of RMS 230v AC
A = 325.27
# approximate value of pi
pi = 3.141592653659
# frequency of the AC line voltage in Hz
```

```python
# Number of samples, defining how many probes we take along time axis
samples = 15
# first create an empty list (empty array) for the time axis
time = []
# using a for loop over a range of x values, generate points on 
# the x time axis in form of a list (array) starting from 0?
# stoppping at the maxinum samples, with step of one, and append
# to the list of time points which was originally empyt
for x in range(0, samples, 1):
  time.append(x)
```

```python
print(time)
```

```python
# first create an empty list (empty array) for the yValue axis
yValue = []
# using a for loop, go over the list of time values
# compute the amplitude of the sine wave at each sample point
# using the general formula from above, substituting in the values
for t in time:
    y = A * math.sin(2 * pi * f * (t / samples))
    # append the newly calculted y value to the yValue list
    yValue.append(y)
```
